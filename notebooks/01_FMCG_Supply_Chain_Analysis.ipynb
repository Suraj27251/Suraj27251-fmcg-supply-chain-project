## 5. OTIF (ON TIME IN FULL) ANALYSIS

**OTIF** is the most critical KPI in supply chain management.

- **On-Time Delivery %**: Orders delivered on or before agreed delivery date
- **In-Full Delivery %**: Orders delivered with full ordered quantity
- **OTIF %**: Orders meeting BOTH conditions (On-Time AND In-Full)

Industry targets typically range from 80-95% OTIF.
